### Lesson HEllOW WORLD AGENT

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python

# This is the state of the node the way it will store data

# The class named AgentState is an State Schema
class AgentState(TypedDict):
    message:str

# This is node here
def greeting_node(state:AgentState)->AgentState:
    """Simple Node that adds a Greeting Message to the State

    Args:
        state (AgentState): _description_

    Returns:
        AgentState: _description_
    """

    state['message'] = "Hey  "+state['message']+", how is your day going?"

    return state
```

```python
# Using StateGraph To build and compile 
graph = StateGraph(AgentState)

graph.add_node("greeter",greeting_node)

graph.set_entry_point("greeter")
graph.set_finish_point("greeter")

app = graph.compile()
```

```python
from IPython.display import Image, display 
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message":"Bob"})
```

```python
result["message"]
```

### Exercise for Graph 1

**TASK** Create a Personalized Compliment Agent using LangGraph!

**Input**:{'name',:'Bob'}<br>
**OUTPUT**: "Bob, Your're doing an amazing job learning LangGraph!"

```python
class AgentState(TypedDict):
    message:str

def complement_node(state:AgentState)->AgentState:
    """It will add an Complement To the message sent by the user
    Args:
        state (AgentState): _description_

    Returns:
        AgentState: _description_
    """
    state['message'] = 'Hey, '+state['message'] + " you're doing an amazing job learning langGraph"
    return state


graph = StateGraph(AgentState)
graph.add_node('complement',complement_node)
graph.set_entry_point('complement')
graph.set_finish_point('complement')

app = graph.compile()

from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
    
```

```python
result = app.invoke({'message':'Bob'})
```
